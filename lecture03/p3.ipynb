```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 3, 3, 4, 4, 5, 5, 6, 6, 7, 8, 9, 10])
y = np.array([63, 68, 72, 32, 87, 77, 98, 76, 83, 89, 94, 97, 99, 100, 100])

m = 2.0
c = 50.0
alpha = 0.005
epochs = 2000

n = len(x)
loss_history = []

for i in range(epochs) :
    y_pred = m * x + c
    error = y - y_pred

    loss = np.mean(error ** 2)
    loss_history.append(loss)
    
    dm = -2 * np.mean(x * error)
    dc = -2 * np.mean(error)

    m -= alpha * dm
    c -= alpha * dc

print("기울기 m = ", m)
print("절편 c = ", c)
print(f'학습된 모델 : y = {m:.2f}x + {c:.2f}')
```

```python
x_line = np.linspace(min(x), max(x), 100)
y_line = m * x_line + c

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color = 'blue', label = 'Actual Data')
plt.plot(x_line, y_line, color = 'red', label=f'Learned Line: y = {m:.2f}x + {c:.2f}')
plt.title('Linear Regression Trained with Gradiend Descent')
plt.xlabel('Study Hours')
plt.ylabel("Exam Score")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(loss_history, color = 'green')
plt.title('MSE (Loss) Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.grid()
plt.show()
```
